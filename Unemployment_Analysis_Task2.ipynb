```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import piplite
await piplite.install('seaborn')
import seaborn as sns
```

```python
df = pd.read_csv("Unemployment in India.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'], color='orange')
plt.title("Unemployment Rate in India During Covid-19")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=df)
plt.xticks(rotation=90)
plt.title("Region-wise Unemployment Rate in India")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="YlOrRd")
plt.title("Correlation Heatmap")
plt.show()

```

```python
sns.boxplot(x='Area', y='Estimated Unemployment Rate (%)', data=df)
plt.title("Urban vs Rural Unemployment Rate")
plt.show()
```
